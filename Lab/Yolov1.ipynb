```
from tensorflow.keras.layers import Conv2D,MaxPooling2D,LeakyReLU,BatchNormalization,Dense,Flatten,Input
from tensorflow.keras.models import Model
```

```
#input 
input_layer = Input(shape=(448,448,3))
activation_leaky = LeakyReLU(alpha=0.1)
# Convolution block 0
out_0      = Conv2D(64, kernel_size=(7,7), strides=(2,2), name='conv_0',activation=activation_leaky,padding="same")(input_layer)
pooling_0  = MaxPooling2D(pool_size=(2,2), strides=2, name='pooling_0')(out_0)
# Convolution block 1
out_1      = Conv2D(192, kernel_size=(3, 3), strides=(1, 1), name='conv_1',activation=activation_leaky,padding="same")(pooling_0)
pooling_1  = MaxPooling2D(pool_size=(2,2), strides=2, name='pooling_1')(out_1)
# Convolution block 2
out_2      = Conv2D(128, kernel_size=(1, 1), strides=(1, 1), name='conv_2',activation=activation_leaky,padding="same")(pooling_1)
out_3      = Conv2D(256, kernel_size=(3, 3), strides=(1, 1), name='conv_3',activation=activation_leaky,padding="same")(out_2)
out_4      = Conv2D(256, kernel_size=(1, 1), strides=(1, 1), name='conv_4',activation=activation_leaky,padding="same")(out_3)
out_5      = Conv2D(512, kernel_size=(3, 3), strides=(1, 1), name='conv_5',activation=activation_leaky,padding="same")(out_4)
pooling_2  = MaxPooling2D(pool_size=(2,2), strides=2, name='pooling_2')(out_5)
# Convolution block 3
out_6      = Conv2D(256, kernel_size=(1, 1), strides=(1, 1), name='conv_6',activation=activation_leaky,padding="same")(pooling_2)
out_7      = Conv2D(512, kernel_size=(3, 3), strides=(1, 1), name='conv_7',activation=activation_leaky,padding="same")(out_6)
out_8      = Conv2D(256, kernel_size=(1, 1), strides=(1, 1), name='conv_8',activation=activation_leaky,padding="same")(out_7)
out_9      = Conv2D(512, kernel_size=(3, 3), strides=(1, 1), name='conv_9',activation=activation_leaky,padding="same")(out_8)
out_10     = Conv2D(256, kernel_size=(1, 1), strides=(1, 1), name='conv_10',activation=activation_leaky,padding="same")(out_9)
out_11     = Conv2D(512, kernel_size=(3, 3), strides=(1, 1), name='conv_11',activation=activation_leaky,padding="same")(out_10)
out_12     = Conv2D(256, kernel_size=(1, 1), strides=(1, 1), name='conv_12',activation=activation_leaky,padding="same")(out_11)
out_13     = Conv2D(512, kernel_size=(3, 3), strides=(1, 1), name='conv_13',activation=activation_leaky,padding="same")(out_12)
out_14     = Conv2D(512, kernel_size=(1, 1), strides=(1, 1), name='conv_14',activation=activation_leaky,padding="same")(out_13)
out_15     = Conv2D(1024,kernel_size=(3, 3), strides=(1, 1), name='conv_15',activation=activation_leaky,padding="same")(out_14)
pooling_3  = MaxPooling2D(pool_size=(2,2), strides=2, name='pooling_3')(out_15)
# Convolution block 4
out_16     = Conv2D(512, kernel_size=(1, 1), strides=(1, 1), name='conv_16',activation=activation_leaky,padding="same")(pooling_3)
out_17     = Conv2D(1024, kernel_size=(3, 3), strides=(1, 1), name='conv_17',activation=activation_leaky,padding="same")(out_16)
out_18     = Conv2D(512, kernel_size=(1, 1), strides=(1, 1), name='conv_18',activation=activation_leaky,padding="same")(out_17)
out_19     = Conv2D(1024, kernel_size=(3, 3), strides=(1, 1), name='conv_19',activation=activation_leaky,padding="same")(out_18)
out_20     = Conv2D(1024, kernel_size=(3, 3), strides=(1, 1), name='conv_20',activation=activation_leaky,padding="same")(out_19)
out_21     = Conv2D(1024, kernel_size=(3, 3), strides=(2, 2), name='conv_21',activation=activation_leaky,padding="same")(out_20)
# Convolution block 5
out_22     = Conv2D(1024, kernel_size=(3, 3), strides=(1, 1), name='conv_22',activation=activation_leaky,padding="same")(out_21)
out_23     = Conv2D(1024, kernel_size=(3, 3), strides=(1, 1), name='conv_23',activation=activation_leaky,padding="same")(out_22)
flatten = Flatten(name='Flatten')(out_23)
fc_0 = Dense(units=512,name='fc_0')(flatten)
fc_1 = Dense(units=4096,name='fc_1')(fc_0)
fc_2 = Dense(units=7*7*30, name='fc_2')(fc_1)
```

```
model = Model(inputs=input_layer, outputs=fc_2)
model.summary()
```
